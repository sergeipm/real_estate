### Шаг 2. Предобработка данных

```python

df['floors_total'].hist(bins = 30)
plt.xlim(0,30)
plt.xlabel('Количество этажей')
plt.title('Этажность домов')
```

Большинство домов ниже 30 этажей. Самыми распространёнными можно назвать дома до 10 этажей. 

```python
print(df[['floor', 'floors_total']].describe())
df['floors_total'].hist()
df['floor'].hist()
```

Этажность дома в среднем в два раза ниже указанного этажа, пропуски можно заполнить опираясь на эту закономерность

```python
# для квартир, высота потолков которых указана меньше чем 2.2. метра,  значения будут приведеены к 2.2 метра
# Кваритры с потолками больше 6 метров также будут приведены к 6 метрам
# Так в данных останется информация о том высокие или низкие потолки, а сильные аномалии не повлияют на общую выборку
def ceiling_fix(row):
    ceiling = row['ceiling_height']
    if ceiling < 2.2:
        return 2.2
    if ceiling > 6:
        return 3.2
    return ceiling

# применяем функцию
df['ceiling_height'] = df.apply(ceiling_fix, axis = 1)



# заполняем пропуски этажности домов
def floors_fill(row):
    floors_total = row['floors_total']
    floor = row['floor']
    if floors_total < 61:
        return floors_total
    else:
        floors = 2 * floor
        return floors
    
# применяем функцию
df['floors_total'] = df.apply(floors_fill, axis = 1)

# Пропуски для апартаментов заполняем отрицательными значениями, так как это редкий тип недвижимости
df['is_apartment'] = df['is_apartment'].fillna(False)
# балконы заполняем значением 0, что означает, что балкона нет
# если пользователь не указал эту информацию, скорей всего балкона нет, если данные пропали по другим причинам, отстутствие 
# балкона не введёт покупателя в заблуждение, не будет расчитывать на то чего нет.
df['balcony'] = df['balcony'].fillna(0)
```


1. Высота потолков. Трудно предствавить квартиру с потолками меньше двух метров, по нормам даже в коридорах потолки должны быть не меньше 2.1. По средним значениям высоты потолков значения меньше чем 2.2 метра были заменены на 2.2. Потолки больше 6 метров тоже очень странная вещь, похоже на аномалию.  С высотой потолка 3.2 высота будет более чем достаточная. Количество выбросов всего несколько штук и на статистику это не повлияет. Заполнять пропущенные значения потолков не стоит, слишком велик шанс ошибиться и испортить статистику. Пропущенных значений почти половина.
2. Этажность дома. Пропущено всего 30 значений, их можно заполнить средними значениями. По таблице видна связь этажа квартиры с этажностью дома, этаж в два раза меньше, пропуски так и заполнены.
3. Апартаменты, этот вариант встречается очень редко, для заполненных значений всего несколько десятков людей указали такой тип жилья. Если заполнять пропущенные значения, то только отметкой 'не является апартаментами'.
4. Площадь кухни и жилая площадь. Их можно заполнить средними значениями, высчитав коэффициент, но такие параметры для кого то могут быть важны и вводить в заблуждение не стоит. Пропуски остаются.
5. Количество дней публикации. Для незаполненных значений из 3000 9/10 относятся к 2018 и 2019 годам. Обьявления свежие, может даже связано с тем что они еще публикуются и произошла какая то ошибка при выгрузке. Заполнять нечем.
6. Название населенного пункта не заполнено в 50 случаях, почти для всех есть расстояния от центра, при необходимости можно попробовать заполнить пропуски посмотрев соответствует ли соотношение цены и расстояния от центра Питерским.
7. Согласно описанию заданию пропуски балконов заполнены нолями, но до заполнения в среднем каждая вторая квартира была с балконом, после такого заполнения статистика изменится и не факт,что в правильную сторону.


Во всех случаях пропусков важно в итоге разобраться почему они произошли, каждая квартира может быть заполнена всеми параметрами, какие то нужно автоматизировать в большей степени, тогда этажность дома как сейчас будет подтягиваться из типа дома по адресу, а расстояния до важным для проживания мест будут отображаться для всех квартир.

Пропуски связаны, скорей всего, с человеческим фактором, люди не хотят или не знают как заполнить некоторые поля при размещение объявления и оставляют их пустыми.


### Дополнительные параметры для оценки данных


Чтобы изучить влияние различных факторов на стоимость квартиры добавим новые параметры на основе существующих

```python
# стоимость квадратного метра
df['price_per_meter'] = (df['last_price'] / df['total_area'])
# приведение данных о размещении объявления к дате
df['first_day_exposition'] = pd.to_datetime(df['first_day_exposition'],format='%Y-%m-%dT')
# день недели
df['day'] = df['first_day_exposition'].dt.weekday
# месяц
df['month'] =df['first_day_exposition'].dt.month
# год
df['year'] = df['first_day_exposition'].dt.year

# для оценки влияния этажа на стоимость выделим три категории
def floor_group(row):
    floor = row['floor']
    floors_total = row['floors_total']    
    if floor == 1:
        return 'первый'
    if floor == floors_total and floor != 1:
        return 'последний'
    return 'другой'

# применение функции для нахождения типа этажа
df['floor_group'] = df.apply(floor_group, axis = 1)
# нахождение коэффициентов жилой площади и площади кухни
df['livig_area_ratio'] = df['living_area'] / df['total_area']
df['kitchen_area_ratio'] = df['kitchen_area'] / df['total_area']
```
